### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize graph
plt.figure(figsize=(5,3))

# setting dpi = 300
#plt.figure(figsize=(5,3), dpi=300)  # in this case, you have 1500 dpi x 900 dpi graph

plt.plot(x,y, label='2x', color='blue', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

## line number two
## using different styles in different segment
x2 = np.arange(0, 4.5, 0.5)   # note 4.5 is excluded 
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')
print(x2)

# Title and labels
plt.title("This is the title", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# tickmarks
plt.xticks([0, 1, 2, 3])
#plt.yticks([0, 2, 4, 6, 8, 10])

# add legend
plt.legend()

# save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

plt.show()  # get rid of the line [<matplotlib.lines.Line2D at 0x118f08ef0>]
```

### Bar chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

# different patterns of bar
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))   # pop out index of the object to be removed from the list.
        
plt.show()
```
